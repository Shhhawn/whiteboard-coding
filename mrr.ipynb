“第一个”真正相关的物品，排在什么位置。

计算逻辑：RR（Reciprocal Rank，倒数排名）：对于某一次查询（或某个用户），找到第一个相关物品的所在排名位置 $R$。它的 RR 分数就是 $\frac{1}{R}$。
- 如果第 1 个就是相关的，得分 $\frac{1}{1} = 1.0$。
- 如果第 3 个才是相关的，得分 $\frac{1}{3} \approx 0.33$。
- 如果一直没找到，得分 $0$。

MRR（Mean RR）：把所有用户（或所有查询）的 RR 分数加起来，求个平均值。$$MRR = \frac{1}{|Q|} \sum_{i=1}^{|Q|} \frac{1}{R_i}$$

```python
def calculate_mrr(y_trues):
    """
    计算MRR
    :param y_trues: List[List[int]], 每个子列表代表一个查询的真实相关性标签(0或1)，
                    且已经按模型的预测打分从高到低排序好了。
                    例如: [[0, 0, 1], [1, 0, 0], [0, 0, 0]]
    :return: float, MRR 得分
    """
    if not y_trues:
        return 0.0

    rr_sum = 0
    for y_true in y_trues:
        for idx, label in enumerate(y_true):
            if label > 0:
                rr_sum += 1 / (idx + 1)
                break
        
    return rr_sum / len(y_trues)

print("MRR:", calculate_mrr([[0, 0, 1], [1, 0, 0], [0, 0, 0]]))
```
