```python
import random
import math

def sigmoid(x):
    x = max(-500, min(x, 500))
    return 1.0 / (1.0 + math.exp(-x))

def compute_loss(y_true, pred):
    eps = 1e-15
    pred = max(eps, min(1-eps, pred))
    return -(y_true * math.log(pred) + (1 - y_true) * math.log(1 - pred))

def forward(x_sample, weights, bias):
    # output = sig( w * x + b )
    z = 0.0
    for i in range(len(x_sample)):
        z += x_sample[i] * weights[i]
    z += bias
    return sigmoid(z)

def backward_and_update(x_sample, pred, y_true, weights, bias, lr=0.0001):
    compute_loss(y_true, pred)
    error = pred - y_true
    for i in range(len(weights)):
        gradient_w = error * x_sample[i]
        weights[i] -= lr * gradient_w

    gradient_b = error
    bias -= lr * gradient_b

    return weights, bias


def train():
    NUM_FEATS = 5
    NUM_SAMPLES = 10
    EPOCHS = 100
    LR = 0.1
    X = [[random.uniform(-1, 1) for _ in range(NUM_FEATS)] for _ in range(NUM_SAMPLES)]
    Y = [random.randint(0,1) for _ in range(NUM_SAMPLES)]

    weights = [random.uniform(-0.01, 0.01) for _ in range(NUM_FEATS)]
    bias = 0.0

    for epoch in range(EPOCHS):
        total_loss = 0

        for i in range(NUM_SAMPLES):
            pred = forward(X[i], weights, bias)
            loss = compute_loss(Y[i], pred)
            total_loss += loss

            wieghts, bias = backward_and_update(X[i], pred, Y[i], weights, bias, lr=LR)

        if (epoch + 1) % 10 == 0:
            print(f"Epoch: {epoch}, total loss: {total_loss}")

train()
    



```

```python
import random
import math

NUM_FEATS = 8
NUM_SAMPLES = 10
NUM_EPOCHS = 500
PRINT_FREQ = 10

X = [[random.uniform(-1, 1) for _ in range(NUM_FEATS)] for _ in range(NUM_SAMPLES)]
Y = [random.randint(0, 1) for _ in range(NUM_SAMPLES)]
W = [random.uniform(-0.01, 0.01) for _ in range(NUM_FEATS)]
B = 0.0

def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

def linear(X, weights, bias):
    assert len(X) == len(weights), "linear input shape is not equal to weights shape"
    output = 0.0
    for x, w in zip(X, weights):
        output += w * x

    output += bias
    return output

def forward(X, weights, bias):
    linear_output = linear(X, weights, bias)
    return sigmoid(linear_output)

def calculate_loss(pred, y_true):
    return - (y_true * math.log(pred) + (1 - y_true) * math.log(1 - pred))


def backward_and_update(X, y_true, pred, weights, bias, lr=0.0001):
    # dL / dsig * dsig / dz * dz / dw 
    # (sig - y) / sig(1-sig) * sig*(1 - sig) * x

    dz = pred - y_true

    for i in range(len(weights)):
        grad_w = dz * X[i]
        weights[i] -= lr * grad_w
    
    grad_b = dz
    bias -= lr * grad_b

    return weights, bias


for epoch in range(NUM_EPOCHS):
    total_loss = 0
    for i in range(len(X)):
        sample = X[i]
        label = Y[i]
        pred = forward(sample, W, B)
        loss = calculate_loss(pred, label)
        total_loss += loss
        W, B = backward_and_update(sample, label, pred, W, B, lr=0.01)

    if (epoch + 1) % PRINT_FREQ == 0:
        print(f"epoch: {epoch + 1}, total loss: {total_loss}")

    
```
