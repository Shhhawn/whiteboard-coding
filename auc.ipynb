## AUC 计算 (Mann-Whitney U 统计量法):

总正样本 ($P$) = 2, 总负样本 ($N$) = 2

我们需要看每个正样本排在了多少个负样本前面：

- 正样本 (0.2): 排在 0 个负样本前面 (它是最小的)。得分 = 0。
- 正样本 (0.7): 排在 2 个负样本 (0.3, 0.5) 前面。得分 = 2。

总得分 = 0 + 2 = 2。

$AUC = 总得分 / (P \times N) = 2 / (2 \times 2) = 0.5$。

```python
import itertools
def calculate_auc(pred_list: list, click_list: list):
    num_pos = sum(click_list)
    num_neg = len(click_list) - num_pos
    pred_click = sorted(zip(pred_list, click_list), key=lambda x:x[0])
    neg_for_now = 0
    auc = 0

    for pred, pairs in itertools.groupby(pred_click, key=lambda x:x[0]):
        pos_cnt = 0
        pair_cnt = 0
        for _, cl in pairs:
            pair_cnt += 1
            if cl:
                pos_cnt += 1
        neg_cnt = pair_cnt - pos_cnt
        auc += pos_cnt * neg_for_now + pos_cnt * neg_cnt / 2.0

        neg_for_now += neg_cnt

    return auc / (num_neg * num_pos)

pred = [0.2, 0.3, 0.7, 0.5]
label = [1, 0, 1, 0]

print(calculate_auc(pred, label))   # 0.5 
```
